```python
import re
import pandas as pd
from pathlib import Path

# =========================================================
# 1. 路径
# =========================================================
input_file = Path("/mnt/zzbnew/peixunban/zhanghebin/CIMA_multiomics_regulation/data/raw/CIMA/meta/招募汇总467例-整合问卷&临床信息-20220109 -无姓名.xlsx")
output_dir = Path("/mnt/zzbnew/peixunban/zhanghebin/CIMA_multiomics_regulation/data/processed/CIMA/meta")
output_dir.mkdir(parents=True, exist_ok=True)

output_clean_csv = output_dir / "CIMA_meta_sheet2_clean.csv"
output_analysis_csv = output_dir / "CIMA_meta_sheet2_analysis_ready.csv"

# =========================================================
# 2. 工具函数
# =========================================================
def excel_col_to_idx(col):
    col = col.upper()
    idx = 0
    for c in col:
        idx = idx * 26 + (ord(c) - ord("A") + 1)
    return idx - 1

def get_col_range(start_col, end_col):
    return list(range(excel_col_to_idx(start_col), excel_col_to_idx(end_col) + 1))

def clean_string(x):
    if pd.isna(x):
        return pd.NA
    s = str(x).strip()
    s = s.replace("　", " ")
    s = re.sub(r"\s+", " ", s)
    return s if s != "" else pd.NA

# =========================================================
# 3. 读取 sheet2
# =========================================================
df_raw = pd.read_excel(input_file, sheet_name=1, header=None)
print("raw shape:", df_raw.shape)

# 如果后面发现首行不是样本，把 3 改成 2
start_row = 3
df = df_raw.iloc[start_row:, :].copy().reset_index(drop=True)

# =========================================================
# 4. 列名定义
# =========================================================
base_cols = {
    "I": "CIMA_ID",
    "G": "体检编码",
    "H": "BGE编号",
    "J": "性别",
    "K": "年龄",
    "F": "出生日期",
    "A": "备注",
}

questionnaire_cols = [
    "籍贯",
    "身高_cm",
    "体重",
    "体重指数",
    "民族",
    "血型",
    "是否处于经期",
    "是否接种过以下疫苗",
    "两周内是否有以下症状",
    "近一个月内心情",
    "最近是否有以下感觉",
    "是否患有以下疾病",
    "手术史",
    "手术部位",
    "近两周内使用过的药物",
    "过敏经历_过敏原",
    "家族成员三代内是否有2人及以上患相同疾病",
    "吸烟",
    "烟龄",
    "接触二手烟",
    "每周的酒精摄入",
    "饮酒频率",
    "饮食习惯",
    "饮食频率",
    "睡眠情况",
    "睡眠时长",
    "日常工作或生活中主要身体姿势",
    "经常进行什么强度的运动",
    "运动习惯",
    "平均每天运动时长",
    "问卷备注",
]

biochem_value_cols = [
    "γ_谷氨酰转肽酶",
    "白蛋白",
    "白蛋白_球蛋白_A_G",
    "胆固醇",
    "低密度脂蛋白胆固醇",
    "甘油三酯",
    "高密度脂蛋白胆固醇",
    "谷丙转氨酶",
    "谷草转氨酶",
    "谷草转氨酶_谷丙转氨酶",
    "肌酐",
    "间接胆红素_IBIL",
    "尿素_UR",
    "尿酸",
    "球蛋白_GLOB",
    "血糖",
    "直接胆红素_DBIL",
    "总胆红素",
    "总蛋白",
]

biochem_flag_cols = [
    "γ_谷氨酰转肽酶_提示",
    "白蛋白_提示",
    "白蛋白_球蛋白_A_G_提示",
    "胆固醇_提示",
    "低密度脂蛋白胆固醇_提示",
    "甘油三酯_提示",
    "高密度脂蛋白胆固醇_提示",
    "谷丙转氨酶_提示",
    "谷草转氨酶_提示",
    "谷草转氨酶_谷丙转氨酶_提示",
    "肌酐_提示",
    "间接胆红素_IBIL_提示",
    "尿素_UR_提示",
    "尿酸_提示",
    "球蛋白_GLOB_提示",
    "血糖_提示",
    "直接胆红素_DBIL_提示",
    "总胆红素_提示",
    "总蛋白_提示",
    "血液生化是否异常",
    "血液生化备注",
]

ultrasound_cols = [
    "超声检查部位",
    "超声检查结果",
    "超声检查提示",
]

mri_ct_cols = [
    "MRI_CT_str_body_name",
    "MRI_CT_varreport_yin_yang_xing",
    "MRI_CT_varreport_examine_diagnosis",
    "MRI_CT_varreport_examine_impression",
]

assert len(get_col_range("L", "AP")) == len(questionnaire_cols)
assert len(get_col_range("AQ", "BI")) == len(biochem_value_cols)
assert len(get_col_range("BJ", "CD")) == len(biochem_flag_cols)
assert len(get_col_range("CE", "CG")) == len(ultrasound_cols)
assert len(get_col_range("CH", "CK")) == len(mri_ct_cols)

# =========================================================
# 5. 组装 clean 表
# =========================================================
out = pd.DataFrame()

for col_letter, new_name in base_cols.items():
    out[new_name] = df.iloc[:, excel_col_to_idx(col_letter)]

for idx, new_name in zip(get_col_range("L", "AP"), questionnaire_cols):
    out[new_name] = df.iloc[:, idx]

for idx, new_name in zip(get_col_range("AQ", "BI"), biochem_value_cols):
    out[new_name] = df.iloc[:, idx]

for idx, new_name in zip(get_col_range("BJ", "CD"), biochem_flag_cols):
    out[new_name] = df.iloc[:, idx]

for idx, new_name in zip(get_col_range("CE", "CG"), ultrasound_cols):
    out[new_name] = df.iloc[:, idx]

for idx, new_name in zip(get_col_range("CH", "CK"), mri_ct_cols):
    out[new_name] = df.iloc[:, idx]

# 只保留有 CIMA_ID 的样本
out["CIMA_ID"] = out["CIMA_ID"].astype("string").str.strip()
out = out[~out["CIMA_ID"].isna() & (out["CIMA_ID"] != "")].copy()

# 字符清理
for c in out.columns:
    if out[c].dtype == "object" or str(out[c].dtype).startswith("string"):
        out[c] = out[c].map(clean_string)

# 类型处理
out["出生日期"] = pd.to_datetime(out["出生日期"], errors="coerce")
out["年龄"] = pd.to_numeric(out["年龄"], errors="coerce")

numeric_cols = [
    "身高_cm", "体重", "体重指数",
    "γ_谷氨酰转肽酶", "白蛋白", "白蛋白_球蛋白_A_G", "胆固醇",
    "低密度脂蛋白胆固醇", "甘油三酯", "高密度脂蛋白胆固醇",
    "谷丙转氨酶", "谷草转氨酶", "谷草转氨酶_谷丙转氨酶", "肌酐",
    "间接胆红素_IBIL", "尿素_UR", "尿酸", "球蛋白_GLOB",
    "血糖", "直接胆红素_DBIL", "总胆红素", "总蛋白"
]
for c in numeric_cols:
    out[c] = pd.to_numeric(out[c], errors="coerce")

flag_cols = [c for c in out.columns if c.endswith("_提示")] + ["血液生化是否异常"]
for c in flag_cols:
    out[c] = pd.to_numeric(out[c], errors="coerce")

# 性别简单规范化
out["性别"] = out["性别"].replace({
    "男 ": "男", "女 ": "女",
    "male": "男", "female": "女",
    "M": "男", "F": "女"
})

# CIMA_ID 放第一列
front_cols = ["CIMA_ID", "体检编码", "BGE编号", "性别", "年龄", "出生日期"]
rest_cols = [c for c in out.columns if c not in front_cols]
out = out[front_cols + rest_cols].copy()

print("clean shape:", out.shape)
display(out.head())
```

```python
def normalize_code_string(x):
    if pd.isna(x):
        return pd.NA
    s = str(x).strip()
    s = s.replace(" ", "").replace("　", "")
    s = s.replace("，", ",").replace("；", ";")
    return s if s != "" else pd.NA

def parse_single_choice(x, code_map):
    s = normalize_code_string(x)
    if pd.isna(s):
        return {"raw": pd.NA, "code": pd.NA, "label": pd.NA}

    m = re.match(r"^([A-Z])", s)
    code = m.group(1) if m else s
    label = code_map.get(code, pd.NA)

    return {"raw": s, "code": code, "label": label if pd.notna(label) else s}

def parse_multiselect(x, code_map):
    s = normalize_code_string(x)
    if pd.isna(s):
        return {"raw": pd.NA, "codes": pd.NA, "code_list": [], "labels": pd.NA}

    m = re.match(r"^([A-Z]+)", s)
    prefix = m.group(1) if m else ""
    code_list = []
    seen = set()
    for c in prefix:
        if c in code_map and c not in seen:
            code_list.append(c)
            seen.add(c)

    labels = [code_map[c] for c in code_list]
    return {
        "raw": s,
        "codes": "".join(code_list) if len(code_list) > 0 else pd.NA,
        "code_list": code_list,
        "labels": "|".join(labels) if len(labels) > 0 else pd.NA,
    }

def parse_multiselect_with_text(x, code_map):
    s = normalize_code_string(x)
    if pd.isna(s):
        return {"raw": pd.NA, "codes": pd.NA, "code_list": [], "labels": pd.NA, "text": pd.NA}

    m = re.match(r"^([A-Z]+)", s)
    prefix = m.group(1) if m else ""
    text = s[len(prefix):].strip() if m else s

    code_list = []
    seen = set()
    for c in prefix:
        if c in code_map and c not in seen:
            code_list.append(c)
            seen.add(c)

    labels = [code_map[c] for c in code_list]
    return {
        "raw": s,
        "codes": "".join(code_list) if len(code_list) > 0 else pd.NA,
        "code_list": code_list,
        "labels": "|".join(labels) if len(labels) > 0 else pd.NA,
        "text": text if text != "" else pd.NA,
    }

def add_single_choice_columns(df, col, code_map):
    parsed = df[col].apply(lambda x: parse_single_choice(x, code_map))
    parsed_df = pd.DataFrame(parsed.tolist(), index=df.index)
    df[f"{col}_raw"] = parsed_df["raw"]
    df[f"{col}_code"] = parsed_df["code"]
    df[f"{col}_label"] = parsed_df["label"]
    return df

def add_multiselect_columns(df, col, code_map):
    parsed = df[col].apply(lambda x: parse_multiselect(x, code_map))
    parsed_df = pd.DataFrame(parsed.tolist(), index=df.index)

    df[f"{col}_raw"] = parsed_df["raw"]
    df[f"{col}_codes"] = parsed_df["codes"]
    df[f"{col}_labels"] = parsed_df["labels"]

    for code, label in code_map.items():
        safe_label = re.sub(r"[^0-9A-Za-z_\u4e00-\u9fa5]", "_", label)
        df[f"{col}__{code}__{safe_label}"] = parsed_df["code_list"].apply(lambda lst: int(code in lst))
    return df

def add_multiselect_with_text_columns(df, col, code_map):
    parsed = df[col].apply(lambda x: parse_multiselect_with_text(x, code_map))
    parsed_df = pd.DataFrame(parsed.tolist(), index=df.index)

    df[f"{col}_raw"] = parsed_df["raw"]
    df[f"{col}_codes"] = parsed_df["codes"]
    df[f"{col}_labels"] = parsed_df["labels"]
    df[f"{col}_text"] = parsed_df["text"]

    for code, label in code_map.items():
        safe_label = re.sub(r"[^0-9A-Za-z_\u4e00-\u9fa5]", "_", label)
        df[f"{col}__{code}__{safe_label}"] = parsed_df["code_list"].apply(lambda lst: int(code in lst))
    return df
```
